```python
import os
import re
from PIL import Image
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import cv2
```

```python
image_ids = [(file, int(re.match(r'(\d+)', file).group(1))) for file in os.listdir('data/valid')]
image_ids
```

```python
test_file = image_ids[0]
```

```python
img_path = os.path.join('_data/valid', test_file[0])
img_path
```

```python
with Image.open(img_path) as image:
    image = image.convert('RGB')
    image_arr = np.array(image)
    image_arr_500 = np.array(image.resize((500, 500)))

    f, axarr = plt.subplots(1,2)
    axarr[0].imshow(image_arr)
    axarr[1].imshow(image_arr_500)
    f.show()
```
